# PneumoniaMNIST — exploration

EDA: split sizes, label counts, and sample images. Train models with `python training/train.py` from the project root.

```python
import sys
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np

ROOT = Path.cwd().resolve()
if not (ROOT / "data" / "dataset.py").exists():
    ROOT = ROOT.parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

from medmnist import PneumoniaMNIST

train_ds = PneumoniaMNIST(split="train", download=True)
val_ds = PneumoniaMNIST(split="val", download=True)
test_ds = PneumoniaMNIST(split="test", download=True)

print("Train:", len(train_ds), "Val:", len(val_ds), "Test:", len(test_ds))

labels = [int(np.asarray(train_ds[i][1]).squeeze()) for i in range(len(train_ds))]
u, c = np.unique(labels, return_counts=True)
print("Train label counts:", dict(zip(u.tolist(), c.tolist())))

fig, axes = plt.subplots(1, 4, figsize=(10, 2.5))
for i in range(4):
    img, y = train_ds[i]
    yi = int(np.asarray(y).squeeze())
    axes[i].imshow(np.asarray(img), cmap="gray")
    axes[i].set_title("Pneumonia" if yi else "Normal")
    axes[i].axis("off")
plt.tight_layout()
plt.show()
```
